```python
import torch
from torch.autograd import Variable
import torch.nn as nn
import generator
import sympy

model = nn.Sequential(
    nn.Conv1d(4, 5, 1),
    nn.ReLU(),
    nn.Conv1d(5, 1, 1)
)

loss_fn = torch.nn.MSELoss(size_average=False)

i = iter(generator.FourF(42))
learning_rate = 1e-4
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
for _ in range(200):
    f, x = next(i)
    f = sympy.lambdify(sympy.abc.x, f)
    
    pred = model(Variable(torch.Tensor([list(map(lambda e: [e], x))])))
    
    loss = loss_fn(f(x[0] - pred), Variable(torch.Tensor([0]), requires_grad=False))
    print(loss.data[0])
    
    optimizer.zero_grad()       
                   
    loss.backward()
    
    optimizer.step()

    
```

```python
%matplotlib inline

import itertools
import matplotlib.pyplot as py
import numpy as np

f, x = next(i)
f_lam = sympy.lambdify(sympy.abc.x, f)

f_x_data = list(np.arange(-10, 10, 0.1))
f_y_data = [f_lam(i) for i in f_x_data]

def pred(start_x):
    pred_dx = model(Variable(torch.Tensor(generator.FourF.get_data(f, start_x))))
    print(pred_dx.data[0])
    next_x = start_x + pred_dx.data[0]
    yield (next_x, f_lam(next_x))
    yield from pred(next_x)
    
pred_generator = iter(pred(5))
print(list(itertools.islice(pred_generator, 10)))
pred_x, pred_y = list(zip(*list(itertools.islice(pred_generator, 10))))

py.plot(f_x_data, f_y_data)
py.plot(pred_x, pred_y, '-')

```
